```python
df = pd.read_csv(r"data.csv")
df = df.drop(df.columns[[0]], axis = 1)
print(df.size)
df.head()
```

```python
ax = plt.gca()
df.plot(kind = 'line', x = 'C_Rate', y = 'Voltage', ax = ax, color = 'red')
df.plot(kind = 'line', x = 'C_Rate', y = 'Predicted_Voltage', ax = ax, color = 'blue')
plt.show()
```
